```python
import numpy as np
def calculate_fov(f, w):
    fov_rad = 2 * np.arctan(w/(2*f))
    fov_deg = np.rad2deg(fov_rad)
    return fov_deg
```

```python
import matplotlib.pyplot as plt
def plot_fov(w_1, w_2):
    f = np.arange(10, 200, 1)

    fov1 = calculate_fov(f, w_1)
    fov2 = calculate_fov(f, w_2)

    plt.figure(figsize=(10, 6))
    plt.plot(f, fov1, label=f"Sensor Width: {w_1} mm")
    plt.plot(f, fov2, label=f"Sensor Width: {w_2} mm")
    plt.xlabel("Focal Length (mm)")
    plt.ylabel("Field of View (deg)")
    plt.grid(True)
    plt.legend()
    plt.show()

plot_fov(7, 36)
```
